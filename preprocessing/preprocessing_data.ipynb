### Variables
Looking at those 40 variables on our dataset, it is possible to notice that some of them correspond to correlated situations. As an example, we can take the Kills and Deaths of each side: red side kills will be the same as blue side deaths and vice versa.

Separating some variables, we can move forward to prepare our dataframe for future analysis. Also, we are looking to answer some main questions with our data, so it is good to look into those repeated variables.

#### Match Variables - 4
- <font color=orange>blueWins</font> - match winner **(0 - red / 1 - blue)**
- <font color=orange>blueFirstBlood</font> - match first blood **(0 - red / 1 - blue)**
- <font color=blue>blueKills</font> - blue side kills **(red side deaths)**
- <font color=red>redKills</font> - red side kills **(blue side deaths)**

It is important to notice that some variables were left behind due to their correspondence in other variables. We can use as an example the Advantage and Diff variables, that can be seen through the total variables difference.

### Fixing dataframe

```python
df_variables = ["blueWins", "blueFirstBlood", "blueKills", "redKills",
                "blueAssists", "redAssists",
                "blueWardsPlaced", "blueWardsDestroyed", "redWardsPlaced", "redWardsDestroyed",
                "blueEliteMonsters", "blueTowersDestroyed", "redEliteMonsters", "redTowersDestroyed",
                "blueDragons", "blueHeralds", "redDragons", "redHeralds",
                "blueTotalGold", "blueTotalExperience", "redTotalGold", "redTotalExperience", 
                "blueCSPerMin", "redCSPerMin"]
```

```python
main_df = df[df_variables]
```

```python
rename_variables = {"blueWins": "matchWinner", "blueFirstBlood": "matchFirstBlood"}
main_df = main_df.rename(columns=rename_variables)
```
